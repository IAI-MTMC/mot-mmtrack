```python
import yolov5
model = yolov5.load('../checkpoints/head_crowdhuman_yolov5m.pt')
model.conf = 0.2
model.classes = [1]
```

```python
import cv2
img = cv2.imread('data/Screenshot from 2023-03-21 20-11-28.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# img = 'data/Screenshot from 2023-03-21 20-11-28.png'
result = model(img)
result.show()
```
